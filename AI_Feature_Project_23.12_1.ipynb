```python
def create_rows(num=50000):
        mylist = ["yes", "no"]
        output = [{"name":fake.name(),
                   "email":fake.email(),
                   "city":random.choice(['Dsseldorf']),
                   "Days_works_from_office":random.randint(0,5),
                   "Today_date":fake.date_between_dates(date_start=datetime(2023,1,1), date_end=datetime(2023,1,1)),
                   "date_time_booking":fake.date_between_dates(date_start=datetime(2023,1,1), date_end=datetime(2023,1,30)),
                   "Time_Start":random.randint(7,17),
                   "Event":np.random.choice(mylist, p=[0.2, 0.8]),
                   "Weather":random.choice(['Sunny','Cloudy','Windy','Rainy','Foggy','snowing']),
                   "Places_booked":random.randint(1,50),
                   "Total_places":random.randint(50,50)} for x in range(num)]
                   
        return output
```
